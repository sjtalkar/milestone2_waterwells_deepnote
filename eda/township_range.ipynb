# San Joaquin Valley Township-Ranges
This notebook simply explores the San Joaquin Valley Counties and Township-Ranges geospatial data and computes the surface of each Township-Range

Please refer to the [Public Land Survey System](../doc/assets/plss_sanjoaquin_riverbasin.md) documentation for details about the Public Land Survey System (PLS) Township-Range-Section (TRS) system

```python
import sys
sys.path.append('..')
```

```python
import geopandas as gpd
from lib.wsdatasets import WsGeoDataset
from lib.viz import display_data_on_map
```

```python
township_range = WsGeoDataset(input_geofiles=[])
```

Visualize the intersection fo California Counties and the San Joaquin Valley

```python
display_data_on_map(gpd.clip(township_range.ca_counties_df, township_range.sjv_boundaries), feature="COUNTY")
```

Visualize the San Joaquin Valley Township Ranges

```python
display_data_on_map(township_range.sjv_township_range_df)
```

Compute the land surface area of the San Joaquin Valley Township Ranges

```python
township_range_area = township_range.sjv_township_range_df.copy()
township_range_area = township_range_area.to_crs("EPSG:3347")
township_range_area["AREA"] = township_range_area.area
township_range_area
```

```python
township_range_area[["TOWNSHIP_RANGE", "AREA"]].to_csv("../assets/outputs/tr_area.csv", index=False)
```
