```
import  geodataframe_aggregate_functions  as gaf
from geodataframe_visualization_class import NormalizedDataSliderVisualization
```

```
#This is to support large datasets
alt.data_transformers.enable('default', max_rows=None)

shortage_df =  gaf.get_yearly_data( input_file =  "../assets/outputs/shortage_clean.csv", 
                        time_aggregate_column = 'YEAR',
                        year_avg_column_name =  'YEARLY_SHORTAGE_COUNT',
                        col_to_aggr = 'REPORT_DATE',
                        agg_func =  'count',
                        time_value ='2000')
   
township_aggregate_gdf = gaf.get_normalized_data (gdf =shortage_df,
                            geometry_col_list = [ 'TownshipRange', 'COUNTY'],
                            time_aggregate_column = 'YEAR',
                            year_agg_column_name = 'YEARLY_SHORTAGE_COUNT',
                            col_to_agg = 'REPORT_DATE',
                            agg_func =  'count')

```

When visually examining the above, we can clearly notice the reports of shortage increases in more areas in two periods, during the 2014-2015 drought and then again in 2021, another sever drought year.
